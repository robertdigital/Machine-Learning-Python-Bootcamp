```python
import IPython, graphviz, re
from io import StringIO
from IPython.display import Image
import numpy as np
import pandas as pd
import math
from sklearn import tree
from sklearn.datasets import load_boston, load_iris
from collections import defaultdict
```

```python
from dtreeviz.trees import *

regr = tree.DecisionTreeRegressor(max_depth=5, random_state=666)
boston = load_boston()
```

```python
X_train = pd.DataFrame(boston.data, columns=boston.feature_names)
y_train = boston.target
```

```python
regr = regr.fit(X_train, y_train)
```

```python
viz = dtreeviz(regr, boston.data, boston.target, target_name="price", feature_names=boston.feature_names, orientation="TD")
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2, random_state=666)
iris = load_iris()

data = pd.DataFrame(iris.data)
data.columns = iris.feature_names
iris.target_names

clf = clf.fit(data, iris.target)

dtreeviz(clf, data, iris.target, orientation="TD", feature_names=data.columns, target_name="type", class_names=list(iris.target_names))
```
